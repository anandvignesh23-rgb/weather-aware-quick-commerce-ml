# Weather-Aware Quick Commerce Exploration

This notebook inspects the synthetic dataset and validates the expected weather-driven demand and pricing patterns.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

project_root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(project_root / 'data' / 'synthetic_quick_commerce_data.csv', parse_dates=['date'])
data.head()
```

```python
data.shape, data['city'].nunique(), data['product_id'].nunique(), data['date'].nunique()
```

```python
data.groupby('weather_type')[['units_sold', 'delivery_fee']].mean().round(2).sort_values('delivery_fee')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
data.groupby('weather_type')['units_sold'].mean().sort_values().plot.barh(ax=axes[0], title='Demand by weather')
data.groupby('weather_type')['delivery_fee'].mean().sort_values().plot.barh(ax=axes[1], title='Delivery fee by weather', color='orange')
plt.tight_layout()
```
